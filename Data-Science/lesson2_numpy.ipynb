```python
import numpy as np
a = np.arange(0, 4)  # 0 1 2 3
b = np.arange(1, 5)  # 1 2 3 4
print(np.add(a, b))
print(np.subtract(a, b))
print(np.multiply(a, b))
print(np.divide(a, b))
print(np.power(a, b))
```
