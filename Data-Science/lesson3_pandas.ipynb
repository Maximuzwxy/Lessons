```python
import pandas as pd

df = pd.read_csv('scores.csv')

# #显示所有列
# pd.set_option('display.max_columns', None)

# #显示所有行
# pd.set_option('display.max_rows', None)

# #设置value的显示长度为100，默认为50
# pd.set_option('max_colwidth',100) 
df.info()
df["评分人数"].astype("float")
df["评分"].astype("float")
df.isnull().any()
df
```

```python
# df.groupby("评分网站")["评分"].apply(lambda x: x.fillna(x.mean()))
df["评分"] = df["评分"].fillna(df.groupby("评分网站")["评分"].transform("mean"))
df["评分人数"] = df["评分人数"].fillna(df.groupby("评分网站")["评分人数"].transform("min"))
df
```

```python
import pandas as pd

# 创建一个Series对象，指定名称为'A'，值分别为1, 2, 3, 4
# 默认索引为0, 1, 2, 3
series = pd.Series([1, 2, 3, 4], name='A')

# 显示Series对象
print(series)

# 如果你想要显式地设置索引，可以这样做：
custom_index = [1, 2, 3, 4]  # 自定义索引
series_with_index = pd.Series([1, 2, 3, 4], index=custom_index, name='A')

# 显示带有自定义索引的Series对象
print(series_with_index)
```

```python
import pandas as pd

a = [1111, 2222, 3333]

s = pd.Series(a)

print(s)

# 获取索引
index = s.index
print(index)

# 获取值数组
values = s.values
print(values)

# 获取描述统计信息
stats = s.describe()

print(stats)

# 获取最大值和最小值的索引
max_index = s.idxmax()
min_index = s.idxmin()

print(max_index)
print(min_index)

# 其他属性和方法
print(s.dtype)   # 数据类型
print(s.shape)   # 形状
print(s.size)    # 元素个数
print(s.head())  # 前几个元素，默认是前 5 个
print(s.tail())  # 后几个元素，默认是后 5 个
```
